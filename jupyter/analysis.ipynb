```python
import pandas as pd
import pymongo as pm
```

```python
connect_uri = '' # insert database uri here
```

```python
client = pm.MongoClient(connect_uri)
```

```python
db = client.songs
songs = db['songs']
```

```python
rap_songs = list(songs.find({'genre': 'rap'}))
```

```python
def dict_merge(dict1, dict2):
    return(dict2.update(dict1))
```

```python
song1 = {
    'name': rap_songs[0]['name'],
    'genre': rap_songs[0]['genre'],
    'explicit': rap_songs[0]['explicit']
}

song1_ = {**song1, **rap_songs[0]['audio-features']}

song1_
```

```python
rap_list = []
for song in rap_songs:
    song_dict = {**{
    'name': song['name'],
    'genre': song['genre'],
    'explicit': song['explicit']
    }, **song['audio-features']}
    
    rap_list.append(song_dict)
```

```python
rap_df = pd.DataFrame(rap_list)
```
